```python
connection = mysql.connector.connect(
user='root',
password='Sharfoddin@28',
host='localhost',
database='hotelmanagementsystem'
)

# Create a cursor object to execute SQL queries
cursor = connection.cursor()
# Query data from the 'Customers' table
cursor.execute('SELECT * FROM customers')

#After fetching data from the database we are storing it into Pandas DataFrame
customers_df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])

# Query data from the 'Rooms' table
cursor.execute('SELECT * FROM rooms')
rooms_df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])

# Query data from the 'bookings' table
cursor.execute('SELECT * FROM bookings')
bookings_df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])

# Query data from the 'payments' table
cursor.execute('SELECT * FROM payments')
payments_df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])

# Query data from the 'staff' table
cursor.execute('SELECT * FROM staff')
staff_df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])

# Query data from the 'services' table
cursor.execute('SELECT * FROM services')
services_df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])
```

# Visualizing the distribution of room types using a count plot ðŸ‘‡

```python
# Visualize the distribution of room types using a count plot
plt.figure(figsize=(8, 6))  # Set the figure size
ax = sns.countplot(x='room_type', data=rooms_df, palette='Set2', hue='room_type', legend=False)  # Create a count plot
plt.title('Room Type Distribution')  # Set the title of the plot
plt.xlabel('Room Type')  # Label the x-axis
plt.ylabel('Count')  # Label the y-axis
plt.xticks(rotation=0)  # Rotate the x-axis labels if needed

# Display the count values on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()  # Display the plot
```

- <h4 style= 'color: lightblue'>The Room Type Distribution graph illustrates the distribution of different types of rooms available in the hotel. In this graph Suite room has highest priority <br> <br>
<b>Room Types:</b> The x-axis represents the different types of rooms available, such as Single, Double, Suite, or Deluxe.<br><br>
    <b>Count:</b> The y-axis indicates the count or frequency of each room type, representing how many rooms of each type are available or booked.</h4>

# Visualizing the distribution of booking statuses using a bar plot ðŸ‘‡

```python
# Visualize the distribution of booking statuses using a bar plot with values displayed on top of each bar
plt.figure(figsize=(10, 6))  # Set the figure size
status_counts = bookings_df['booking_status'].value_counts()  # Get the count of each booking status
status_counts.plot(kind='bar', color='gold')  # Create a bar plot
plt.title('Booking Status Distribution')  # Set the title of the plot
plt.xlabel('Booking Status')  # Set the label for the x-axis
plt.ylabel('Count')  # Set the label for the y-axis
plt.xticks(rotation=0)  # Rotate x-axis labels if needed

# Display the count values on top of each bar
for i, count in enumerate(status_counts):
    plt.text(i, count + 15, str(count), ha='center', va='top')

plt.show()  # Display the plot
```
